# Setting up for NXS Computational Tutorial

During the upcoming NXS software tutorial we will be going over some widely used computational tools. To save time, it will be helpful to make sure that the following steps are conducted in advance

## Getting the repository

Repositories, or "repos" are essentially folders that contain files. [GitHub](https://github.com/) is a website that hosts repos. 

If you are familiar with the program `git`, then you likely know how to clone a repo. But, if you haven't used git before, you can easily do this just by going to the NXS repo [on github, here](https://github.com/zjmorgan/nxs-computational-tutorial/tree/main). and download it by clicking on the green `Code` button and selecting `Download ZIP`. After downloading, extract it to a convenient location which I'll refer to as `myTutorialPath`. You should now have a copy of the repo at:
   
```myTutorialPath/nxs-computational-tutorial/```

## Installing some software

During the tutorial we will need to use two key pieces of software:

1) git - a version management program
2) conda (or mamba) - an environment and software package management program rolled into one 

If you already have these installed, just jump ahead to the final section. If not, read on...

As is common in the sphere of scientific computing, there are multiple ways to achieve the same end. And this can be confusing. Below are two suggested approach that worked for me (using a mac)

### Option 1: command line with micromamba or miniconda

For our purposes here, `conda` and `mamba` are almost interchangeable. You can choose one or the other but, in my testing, `mamba` was significantly faster than conda (your mileage may vary).

Furthermore, both `conda` and `mamba` are both available in "lightweight" forms called, repsectively `miniconda` and `micromamba`. These lightweight forms are quicker to install as they don't come with pre-installed packages. 

The following instructions describe using `micromamba` (all of the instructions below work for conda too, just replace `micromamba` with `conda`)

1) Next, install [micromamba](https://mamba.readthedocs.io/en/latest/installation.html) following the instructions for your OS. On macOS, I installed by opening a terminal prompt and typing:

```"${SHELL}" <(curl -L micro.mamba.pm/install.sh)```

2)  open a terminal and navigate to the folder containing the repo
   
```cd myTutorialPath/nxs-computational-tutorial/```

3) in the folder you will find a file called `environment.yml` this contains the information that micromamba (or conda etc) needs to build the environment we will use for the tutorial. When created, and activated, this means you will have access to all the software needed.

4) The environment is created with the command:

```micromamba create -f environment.yml```

This will automatically create and environment called `nxs` and download the software you need into it.

5) If you now activate the environment using
   
```micromamba activate nxs```

you should have access to all the sofware you need. `micromamba` not only installs the software packages you need, but allows you to control interdependencies between different packages. if you type

```micromamba list``` you will see not only all the packages installed, but their version numbers.   

You are now ready to go.

### Option 2: more GUI-based

If you're more comfortable with GUI-type interactions, you can alternatively try to install Anaconda Navigator following the installation instructions for you OS [here](https://www.anaconda.com/download) This comes with a bunch of extra goodies including code editors. 

1) Select the `Environments` tab.
2) Click the `+` icon to add a new environment
3) Select the option to import and then in the import from field select the file `environment.yml` in the top directory of the repo
4) Choose a name for the environment (e.g. `nxs`) and hit import. This will create the environment
5) The environment name should now appear in your list of environments.
6) Next to it is a green "play" button. If you click this you will have several options to activate the environment (e.g. in a terminal, in a jupyter notebook or in a python interpreter) 

## Set up if you already have git and conda/mamba installed

If you already have these tools available, set up is easy: 

1) Clone the environment using git:

```git clone https://github.com/zjmorgan/nxs-computational-tutorial.git```

or

```git clone git@github.com:zjmorgan/nxs-computational-tutorial.git```

(if you use SSH)

2) create the environment using the environment.yml in the repo:

```conda create -f environment.yml```

or 

```micromamba create -f environment.yml```

etc.

depending on which environment manager you use.
